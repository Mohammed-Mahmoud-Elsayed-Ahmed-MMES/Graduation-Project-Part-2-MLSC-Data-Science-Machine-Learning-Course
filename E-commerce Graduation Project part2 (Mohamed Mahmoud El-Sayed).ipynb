```python
# Select relevant features for segmentation
X = df_only_numbers[['gender_id', 'city_id', 'customer_id', 'transaction_id', 'branch_id', 'merchant_id']]
```

```python
from sklearn.preprocessing import StandardScaler

# Scale the data 
scaler = StandardScaler()
X = scaler.fit_transform(X)
```

## K-means clustering

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Initialize lists to store metrics
inertia = []
silhouette_scores = []

# Try different numbers of clusters (k)
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    
    # Inertia (within-cluster sum of squares)
    inertia.append(kmeans.inertia_)
    
    # Silhouette score
    silhouette_avg = silhouette_score(X, kmeans.labels_)
    silhouette_scores.append(silhouette_avg)

# Plot the inertia and silhouette score to determine the optimal k
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

# Plot inertia
plt.subplot(1, 2, 1)
plt.plot(range(2, 11), inertia, marker='o')
plt.title('Inertia for Different Cluster Numbers')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')

# Plot silhouette score
plt.subplot(1, 2, 2)
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('Silhouette Score for Different Cluster Numbers')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')

plt.show()
```

```python
# Fit K-Means with the optimal number of clusters 
optimal_k = 6
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42)
df_only_numbers['cluster'] = kmeans_final.fit_predict(X)

# Calculate final silhouette score
silhouette_avg = silhouette_score(X, df_only_numbers['cluster'])
print(f"Silhouette Score for {optimal_k} clusters: {silhouette_avg}")
```

```python
# Analyze the clusters
for cluster_num in range(optimal_k):
    print(f"\nCluster {cluster_num} Analysis:")
    cluster_data = df_only_numbers[df_only_numbers['cluster'] == cluster_num]
    
    # Summary statistics for each cluster
    print(cluster_data[['customer_id', 'transaction_id', 'branch_id', 'city_id', 'gender_id']].describe())
    
    # Percentage of burned vs subscribed coupons in this cluster
    coupon_status_dist = cluster_data['transaction_status'].value_counts(normalize=True) * 100
    print(f"Coupon Status Distribution for Cluster {cluster_num}:\n{coupon_status_dist}")
```

```python
from sklearn.decomposition import PCA

# Fit K-Means model with the optimal number of clusters
optimal_k = 6
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42)
clusters = kmeans_final.fit_predict(X) 

# Apply PCA for 2D visualization
pca = PCA(n_components=2)
pca_result = pca.fit_transform(X)  # Use the scaled X

# Visualize the clusters
plt.figure(figsize=(10, 7))
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=clusters, cmap='viridis', s=50)

# Mark the cluster centroids
centroids = pca.transform(kmeans_final.cluster_centers_)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=200, alpha=0.75, marker='X')

# Titles and labels
plt.title('Customer Segmentation using K-Means and PCA')
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```
