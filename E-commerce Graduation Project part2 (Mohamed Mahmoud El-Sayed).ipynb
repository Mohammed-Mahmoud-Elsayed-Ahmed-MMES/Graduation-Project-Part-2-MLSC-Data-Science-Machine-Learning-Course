## K-means clustering

#### Determining the Optimal Number of Clusters `K` Using Inertia and Silhouette Score

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Initialize lists to store metrics
inertia = []
silhouette_scores = []

# Try different numbers of clusters (k)
for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    
    # Inertia (within-cluster sum of squares)
    inertia.append(kmeans.inertia_)
    
    # Silhouette score
    silhouette_avg = silhouette_score(X, kmeans.labels_)
    silhouette_scores.append(silhouette_avg)

# Plot the inertia and silhouette score to determine the optimal k
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 5))

# Plot inertia
plt.subplot(1, 2, 1)
plt.plot(range(2, 11), inertia, marker='o')
plt.title('Inertia for Different Cluster Numbers')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')

# Plot silhouette score
plt.subplot(1, 2, 2)
plt.plot(range(2, 11), silhouette_scores, marker='o')
plt.title('Silhouette Score for Different Cluster Numbers')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette Score')

plt.show()
```

#### Applying K-Means Clustering with Optimal `k` and Evaluating Silhouette Score

```python
# Fit K-Means with the optimal number of clusters 
optimal_k = 5
kmeans_final = KMeans(n_clusters=optimal_k, random_state=42)
df_only_numbers['cluster'] = kmeans_final.fit_predict(X)

# Calculate final silhouette score
silhouette_avg = silhouette_score(X, df_only_numbers['cluster'])
print(f"Silhouette Score for {optimal_k} clusters: {silhouette_avg}")
```

#### Cluster Analysis: Summary Statistics and Coupon Status Distribution

```python
# Analyze the clusters
for cluster_num in range(optimal_k):
    print(f"\nCluster {cluster_num} Analysis:")
    cluster_data = df_only_numbers[df_only_numbers['cluster'] == cluster_num]
    
    # Summary statistics for each cluster
    print(cluster_data[['customer_id', 'transaction_id', 'branch_id', 'city_id', 'gender_id']].describe())
    
    # Percentage of burned vs subscribed coupons in this cluster
    coupon_status_dist = cluster_data['transaction_status'].value_counts(normalize=True) * 100
    print(f"Coupon Status Distribution for Cluster {cluster_num}:\n{coupon_status_dist}")
```

## DBSCAN (Density-Based Spatial Clustering of Applications with Noise)

#### Using k-Distance Plot for `Epsilon` Selection in DBSCAN Clustering

```python
from sklearn.neighbors import NearestNeighbors

# Choose k as min_samples (commonly 4 or 5)
k = 5
neighbors = NearestNeighbors(n_neighbors=k)
neighbors_fit = neighbors.fit(X)
distances, indices = neighbors_fit.kneighbors(X)

# Sort and plot the distances to the k-th nearest neighbor
distances = np.sort(distances[:, k-1], axis=0)
plt.plot(distances)
plt.title('k-Distance Plot for Epsilon Selection')
plt.xlabel('Data Points')
plt.ylabel(f'{k}-Distance')
plt.show()
```

#### Evaluating DBSCAN with Different Epsilon Values Based on Silhouette Score

```python
from sklearn.cluster import DBSCAN

# Define a range of eps values for testing from the last graph (the values start from where the raise happen)
eps_values = [0.4, 0.5, 0.6, 0.7, 0.8, 0.9]
min_samples = 5  # Based on your previous selection

# Store silhouette scores for comparison
silhouette_scores_dbscan = []

for eps in eps_values:
    # Apply DBSCAN
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)
    labels = dbscan.fit_predict(X)
    
    # Ignore the case where all points are considered noise (-1 label)
    if len(set(labels)) > 1:
        # Calculate silhouette score for evaluation
        silhouette_avg = silhouette_score(X, labels)
        silhouette_scores_dbscan.append((eps, silhouette_avg))
        print(f"Silhouette Score for eps={eps}: {silhouette_avg}")
    else:
        print(f"DBSCAN with eps={eps} resulted in too much noise.")

# Find the best eps value based on the highest silhouette score
best_eps = max(silhouette_scores_dbscan, key=lambda x: x[1])[0]
print(f"\nOptimal eps based on silhouette score: {best_eps}")
```

### 2D Visualization

```python
# Apply PCA for 2D visualization
X_pca_dbscan = pca.fit_transform(X)

# Fit DBSCAN with the best eps value (0.9)
dbscan = DBSCAN(eps=0.9, min_samples=5)
labels_dbscan = dbscan.fit_predict(X)

# Plot the DBSCAN clusters
plt.figure(figsize=(10, 7))
plt.scatter(X_pca_dbscan[:, 0], X_pca_dbscan[:, 1], c=labels_dbscan, cmap='tab20', s=50)

plt.title("DBSCAN Clustering (eps = 0.9, min_samples = 5)")
plt.xlabel("PCA Component 1")
plt.ylabel("PCA Component 2")
plt.colorbar(label='Cluster Label')
plt.show()
```

## Agglomerative (Hierarchical) Clustering

#### Hierarchical Clustering: `Dendrogram Visualization` with Various `Linkage` Methods 

```python
from scipy.cluster.hierarchy import dendrogram, linkage

# Create linkage matrix using different linkage methods
for method in ['ward', 'complete', 'average', 'single']:
    Z = linkage(X, method=method)

    # Plot the dendrogram
    plt.figure(figsize=(10, 7))
    dendrogram(Z)
    plt.title(f'Dendrogram ({method} linkage)')
    plt.xlabel('Data Points')
    plt.ylabel('Distance')
    plt.show()
```

#### Use the Silhouette Score to get the best `n_clusters` and `linkage` combination

```python
from sklearn.cluster import AgglomerativeClustering

# Function to evaluate agglomerative clustering
def evaluate_agglomerative(X, n_clusters_list, linkages):
    best_results = {}
    for n_clusters in n_clusters_list:
        print(f"\nEvaluating for {n_clusters} clusters:\n")
        for linkage in linkages:
            agglom = AgglomerativeClustering(n_clusters=n_clusters, linkage=linkage)
            labels = agglom.fit_predict(X)
            score = silhouette_score(X, labels)
            print(f"Silhouette Score for {linkage} linkage: {score}")
            best_results[(n_clusters, linkage)] = score
    return best_results

# Cluster sizes to test and linkage methods
n_clusters_list = [100, 6, 3, 2]
linkages = ['ward', 'complete', 'average', 'single']

# Run evaluation and store best results
best_agglomerative_results = evaluate_agglomerative(X, n_clusters_list, linkages)

# Find the best combination of n_clusters and linkage method
best_combination = max(best_agglomerative_results, key=best_agglomerative_results.get)
best_n_clusters, best_linkage = best_combination

print(f"\nBest n_clusters and linkage combination: {best_n_clusters} clusters with {best_linkage} linkage.")
print(f"Best silhouette score: {best_agglomerative_results[best_combination]}")
```
